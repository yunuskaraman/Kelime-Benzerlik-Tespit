```python
import pandas as pd
import numpy as np
import nltk
import string
```

```python
file = pd.read_csv('rows.csv')
file
```

```python
list = ['Date received','Sub-product','Sub-issue','Consumer complaint narrative','Company public response',
        'Tags','Consumer consent provided?','Submitted via','Date sent to company','Company response to consumer','Timely response?',
       'Consumer disputed?']
for i in list:
    file.drop([i], axis=1, inplace=True)
file
```

```python
file.dropna(inplace=True)
file
```

```python
sw = set(nltk.corpus.stopwords.words('english'))
sw
```

```python
poin = set(string.punctuation)
```

```python
def remove_punctuation(text):
    text_nopunct = "".join([x for x in text if x not in poin])
    return text_nopunct
```

```python
file['Product'] = file['Product'].apply(lambda x: remove_punctuation(x))
file['Issue'] = file['Issue'].apply(lambda x: remove_punctuation(x))
file
```

```python
def remove_stopWord(text):
    text = nltk.tokenize.word_tokenize(text)
    list = []
    for word in text:
        if word not in sw:
            list.append(word)
    return " ".join(list)
```

```python
processed_docs = file['Product'].map(remove_stopWord)
file['Product'] = processed_docs
```

```python
processed_docs = file['Issue'].map(remove_stopWord)
file['Issue'] = processed_docs
```

```python
file
```

```python
file = file.reset_index(drop=True)
```

```python
file = file.sort_values('Product')
```

```python
file = file.reset_index(drop=True)
```

```python
file
```

```python
dataIssue = file.sort_values('Issue')
dataIssue = dataIssue.reset_index(drop=True)
dataIssue
```

```python
file.to_csv(r'.\product.csv',index=None, header=True)
```

```python
dataIssue.to_csv(r'.\issue.csv',index=None, header=True)
```
