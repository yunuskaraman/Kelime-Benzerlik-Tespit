```python
def issueNumberBy(data):
    issueNumber = data.groupby('Issue').Issue.count()
    return issueNumber
```

```python
def issueNameBy(data):
    issueName = data.groupby('Issue').apply(lambda df: df.Issue.iloc[0])
    return issueName
```

```python
def productNumberBy(data):
    productNumber = data.groupby('Product').Product.count()
    return productNumber
```

```python
def productNameBy(data):
    productName = data.groupby('Product').apply(lambda df: df.Product.iloc[0])
    return productName
```

```python
def word_benzerlik(text,text1):
    if(len(text)>=len(text1)):
        wordB=text
        wordK=text1
    else:
        wordB=text1
        wordK=text
    benzerWord = 0
    for t in range(len(wordK)):
        for t1 in range(len(wordB)):
            if(wordB[t1].lower()==wordK[t].lower()):
                benzerWord +=1
                break
    return benzerWord/len(wordB)*100
    
```

```python
def benzerlik_Oran(name,number,yuzde):
    dfProduct=pd.DataFrame(columns=['Product1','Product2','Benzerlik-Orani','Product1-Sayi','Product2-Sayi'])
    j=0
    for text in range(len(name)):
        product = nltk.tokenize.word_tokenize(name[text])
        for text1 in range(text+1,len(name)):
            product1 = nltk.tokenize.word_tokenize(name[text1])
            sonuc = word_benzerlik(product,product1)
            if(sonuc>=yuzde and number[text1]>1):
                dfProduct.loc[j]=[name[text],name[text1],int(sonuc),int(number[text]),int(number[text1])]
                j+=1
    return dfProduct
```

```python
def benzerlik_Oran1(name,number,yuzde,say1,say2):
    dfProduct=pd.DataFrame(columns=['Product1','Product2','Benzerlik-Orani','Product1-Sayi','Product2-Sayi'])
    j=0
    for text in range(say1,say2):
        product = nltk.tokenize.word_tokenize(name[text])
        for text1 in range(0,len(name)):
            product1 = nltk.tokenize.word_tokenize(name[text1])
            sonuc = word_benzerlik(product,product1)
            if(sonuc>=yuzde and number[text1]>1):
                dfProduct.loc[j]=[name[text],name[text1],int(sonuc),int(number[text]),int(number[text1])]
                j+=1
    return dfProduct
```

```python
def urunleriKarsilastir(productData,yuzde):
    productName = productNameBy(productData)
    productNumber = productNumberBy(productData)
    dfProduct = benzerlik_Oran(productName,productNumber,yuzde)
    return dfProduct
```

```python
def urunleriKarsilastir1(productData,yuzde,say1,say2):
    productName = productNameBy(productData)
    productNumber = productNumberBy(productData)
    dfProduct = benzerlik_Oran1(productName,productNumber,yuzde,say1,say2)
    return dfProduct
```

```python
def issueKarsilastir(issueData,yuzde):
    issueName = issueNameBy(issueData)
    issueNumber = issueNumberBy(issueData)
    dfProduct = benzerlik_Oran(issueName,issueNumber,yuzde)
    return dfProduct
```

```python
def secilenId(Id,yuzde):
    secilenName = productData.loc[productData['Complaint ID']==Id]
    i=ProductNumberBul(secilenName['Product'].item())
    j=0
    productName = productNameBy(productData)
    productNumber = productNumberBy(productData)
    dfProduct=pd.DataFrame(columns=['Product1','Product2','Benzerlik-Orani','Product1-Sayi','Product2-Sayi'])
    for text1 in range(0,len(productName)):
        product = nltk.tokenize.word_tokenize(secilenName['Product'].item())
        product1 = nltk.tokenize.word_tokenize(productName[text1])
        sonuc = word_benzerlik(product,product1)
        if(sonuc>=yuzde):
            dfProduct.loc[j]=[secilenName['Product'].item(),productName[text1],int(sonuc),int(productNumber[i]),int(productNumber[text1])]
            j+=1
    return dfProduct
```

```python
class CPUPainter:
    
    def threading1(self):
        self.df = urunleriKarsilastir1(productData,self.yuzde,0,6)
    def threading2(self):
         self.df1 = urunleriKarsilastir1(productData,self.yuzde,6,12)
    def threading3(self):
         self.df2 = urunleriKarsilastir1(productData,self.yuzde,12,18)
    def __init__(self,yuzde):
        self.yuzde = yuzde
        self.df = pd.DataFrame(columns=['Product1','Product2','Benzerlik-Orani','Product1-Sayi','Product2-Sayi'])
        self.df1 = pd.DataFrame()
        self.df2 = pd.DataFrame()
        t1 = threading.Thread(target=self.threading1)
        t1.start()
        t2 = threading.Thread(target=self.threading2)
        t2.start()
        t3 = threading.Thread(target=self.threading3)
        t3.start()
        
        
    def don(self):
        print(threading.enumerate())
        return pd.concat([self.df,self.df1,self.df2]).reset_index(drop=True)
    def calisanThread(self):
        return threading.active_count()
```
